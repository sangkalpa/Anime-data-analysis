```python
# Anime basic info
anime_info = pd.DataFrame({
    "anime_id": [1, 2, 3, 4],
    "name": ["Naruto", "Bleach", "One Piece", "Death Note"]
})

# Anime ratings
anime_ratings = pd.DataFrame({
    "anime_id": [1, 2, 3, 5],  # notice: 4 is missing, 5 is extra!
    "rating": [8.5, 7.9, 8.1, 9.0]
})
```

```python
pd.merge(anime_info, anime_ratings, on = "anime_id")
```

```python
pd.merge(anime_info, anime_ratings, on = "anime_id", how = "left")
```

```python
pd.merge(anime_info, anime_ratings, on = "anime_id", how = "outer")
```

```python
my_ratings = pd.DataFrame({
    "name": ["Naruto", "Death Note", "Attack on Titan"],
    "my_rating": [9.5, 9.8, 9.2]
})
```
